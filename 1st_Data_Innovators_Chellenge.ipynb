INSIGHT 04

** which age people love discount most on products**

```python
age_bins = [18, 25, 35, 45, 55, 65, 70]
age_labels = ['18-25', '26-35', '36-45', '46-55', '56-65', '65+']
df['Age Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)
```

```python
discount_sensitivity_by_age = df.groupby('Age Group')['Discount Availed'].mean() * 100
print(f"Discount sensitivity by age group :\n{discount_sensitivity_by_age.round(2)}")
```

```python
# make a barplot to visualize the Percentage of Purchases with Discount
plt.figure(figsize=(10, 6))
discount_sensitivity_by_age.plot(kind='bar', color='blue')
plt.title('Discount Sensitivity by Age Group')
plt.ylabel('Percentage of Purchases with Discount (%)')
plt.xlabel('Age Group')
plt.show()
```

This bar chart shows the percentage of purchases made with a discount across different age groups. The 65+ age group is the most discount-sensitive, with over 50% of their purchases using discounts, while younger groups show slightly lower but similar levels of discount usage.
